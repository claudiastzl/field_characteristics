```python
### modify data, and possibly full_bloom, and hit CTRL & Enter

import matplotlib.pyplot as plt
import pandas
import math
import numpy as np

data = pandas.read_csv('C:/Users/Claudia/Downloads/example data daily values_cimis Parlier.csv')
#data = pandas.read_csv('C:/Users/Claudia/Downloads/Verona.csv')
#data = pandas.read_csv('C:/Users/Claudia/Downloads/Durham2014.csv')

full_bloom='1/3/2018'

data['Date1']=pandas.to_datetime(data['Date'])
data['Min Air Temp (C)']=(data['Min Air Temp (F)']-32)/1.8
data['Max Air Temp (C)']=(data['Max Air Temp (F)']-32)/1.8


def gdd(Tmin, Tmax, F_or_C):
    
    if Tmin > Tmax:
        print ('Warning: Tmin > Tmax')
        
    if F_or_C == 'C':
        TL=5
        TU=35
    elif F_or_C =='F':
        TL=41
        TU=95
    
    # first case: temp curve above TU
    if Tmin >= TU:
        return TU - TL
    
    # second case: temp curve completely below TL
    elif Tmax <= TL:
        return 0
    
    # third case: curve below TU and possibly partly below TL
    elif Tmax <= TU:    
        return max(0, 2 / math.pi * ( Tmax - Tmin) + Tmin - TL)
    
    # third case: curve partly above TU
    elif Tmax > TU:

        theta2 = np.arcsin( ( TU-Tmin ) / ( Tmax - Tmin ) ) * 1 / math.pi
        
        area_all = 2 / math.pi * ( Tmax - Tmin ) + Tmin

        area_above_TU = (Tmax - Tmin) / math.pi*(np.cos(theta2) - np.cos(1 - theta2)) + Tmin*(1 - 2*theta2)
        
        return max(0, area_all - area_above_TU - TL)
    
def mygdd(row):
    return gdd(row['Min Air Temp (C)'],row['Max Air Temp (C)'], 'C')

def mygddF(row):
    return gdd(row['Min Air Temp (F)'],row['Max Air Temp (F)'], 'F')

data['gddcol_C']=data.apply(mygdd, axis=1)
data['gddcol_F']=data.apply(mygddF, axis=1)

data_from_Mar1st = data[data['Date1']>=full_bloom].copy()

data_from_Mar1st['gdd_sum_C']=data_from_Mar1st['gddcol_C'].cumsum(axis = 0)
data_from_Mar1st['gdd_sum_F']=data_from_Mar1st['gddcol_F'].cumsum(axis = 0)

print("")
print("These are the summed up gdd values 90 days after full bloom (if taken Celsius or Fahrenheit):")
display(data_from_Mar1st[['Date1', 'gdd_sum_C', 'gdd_sum_F']].head(90).tail(1))
```
